```python
import networkx as nx
import matplotlib.pyplot as plt
G=nx.read_pajek("all_data.net")
G = nx.Graph(G)
print(type(G).__name__)
print(nx.info(G))
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
import numpy as np
plt.title("Histogram")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.hist(nx.degree_histogram(G))
plt.show()
```

```python
#centrality measurements.
betweenness_centrality = nx.betweenness_centrality(G)
sorted(betweenness_centrality.items(),reverse=True,key=lambda x:x[1])[:25]
```

```python
closeness_centrality = nx.closeness_centrality(G)
sorted(closeness_centrality.items(),reverse=True,key=lambda x:x[1])[:50]
```

```python
degree_centrality = nx.degree_centrality(G)
sorted(closeness_centrality.items(),reverse=True,key=lambda x:x[1])[:50]
```
